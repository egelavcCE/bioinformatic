```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

def needleman_wunsch(seq1, seq2, match_score=2, mismatch_score=-1, gap_penalty=-2):
    alignments = pairwise2.align.globalms(seq1, seq2, match_score, mismatch_score, gap_penalty, gap_penalty)
    
    best_alignment = alignments[0]
    print(format_alignment(*best_alignment))
seq1 = "AGACTG"
seq2 = "AGCTG"

needleman_wunsch(seq1, seq2)

```

```python
from Bio.Seq import Seq

def motif_arama(dna_sequence, motif):
    # DNA dizisini BioPython Seq sınıfı olarak tanımla
    dna_seq = Seq(dna_sequence)
    indices = []
    index = dna_seq.find(motif)
    
    # Motifin tekrar eden tüm konumlarını bul
    while index != -1:
        indices.append(index)
        index = dna_seq.find(motif, index + 1)
    
    # Sonuçları ekrana yazdır
    print(f"DNA dizisini girin: {dna_sequence}")
    print(f"Aranacak motifi girin: {motif}")
    print(f"Motif '{motif}' {len(indices)} kez bulundu.")
    print(f"Başlangıç indeksleri: {indices}")

# Kullanıcı girişi
dna_sequence = input("DNA dizisini girin: ")
motif = input("Aranacak motifi girin: ")

# Motif arama fonksiyonunu çağırın
motif_arama(dna_sequence, motif)
```
